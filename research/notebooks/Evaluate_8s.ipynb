```python
from ultralytics import YOLO

# Load the model
model = YOLO("../models/detect/model_detect_all_body.pt")

# Run the evaluation
results = model.val(data="../datasets/dataset_all_body/data.yaml")
```

```python
# Print specific metrics
print("Class indices with average precision:", results.ap_class_index)
print("\nAverage precision for all classes:", results.box.all_ap)
print("\nAverage precision:", results.box.ap)
print("\nAverage precision at IoU=0.50:", results.box.ap50)
print("\nClass indices for average precision:", results.box.ap_class_index)
print("\nClass-specific results:", results.box.class_result)
print("\nF1 score:", results.box.f1)
print("\nF1 score curve:", results.box.f1_curve)
print("\nOverall fitness score:", results.box.fitness)
print("\nMean average precision:", results.box.map)
print("\nMean average precision at IoU=0.50:", results.box.map50)
print("\nMean average precision at IoU=0.75:", results.box.map75)
print("\nMean average precision for different IoU thresholds:", results.box.maps)
print("\nMean results for different metrics:", results.box.mean_results)
print("\nMean precision:", results.box.mp)
print("\nMean recall:", results.box.mr)
print("\nPrecision:", results.box.p)
print("\nPrecision curve:", results.box.p_curve)
print("\nPrecision values:", results.box.prec_values)
print("\nSpecific precision metrics:", results.box.px)
print("\nRecall:", results.box.r)
print("\nRecall curve:", results.box.r_curve) 
```
